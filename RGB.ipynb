RGB Filters - Angelika Baloy DS1002

```python
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image

def red_filter(image):
    red = image.copy()
    red[:, :, 1] = 0  #G
    red[:, :, 2] = 0  #B
    return plt.imshow(red)

def green_filter(image):
    green = image.copy()
    green[:, :, 0] = 0 #R
    green[:, :, 2] = 0 #B
    return plt.imshow(green)

def blue_filter(image):
    blue = image.copy()
    blue[:, :, 0] = 0 #R
    blue[:, :, 1] = 0 #G
    return plt.imshow(blue)
```

Image 1: https://upload.wikimedia.org/wikipedia/commons/thumb/e/ec/Mona_Lisa,_by_Leonardo_da_Vinci,_from_C2RMF_retouched.jpg/1200px-Mona_Lisa,_by_Leonardo_da_Vinci,_from_C2RMF_retouched.jpg

```python
url = input("Enter the URL of an image (must be .jpg): ")
image = np.asarray(Image.open(requests.get(url, stream=True).raw))

plt.imshow(image)
plt.title('Original')

plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
red_filter(image)
plt.title('Red Filter')

plt.subplot(2, 3, 2)
green_filter(image)
plt.title('Green Filter')

plt.subplot(2, 3, 3)
blue_filter(image)
plt.title('Blue Filter')

plt.subplot(2, 3, 4)
rg_array = np.array(image)
rg_array[:,:,2] = 0
plt.imshow(rg_array)
plt.title('Red & Green')

plt.subplot(2, 3, 5)
rb_array = np.array(image)
rb_array[:,:,1] = 0
plt.imshow(rb_array)
plt.title('Red & Blue')

plt.subplot(2, 3, 6)
gb_array = np.array(image)
gb_array[:,:,0] = 0
plt.imshow(gb_array)
plt.title('Green & Blue')

plt.show()

```
